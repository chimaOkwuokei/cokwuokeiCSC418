```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def enhancement_techniques(technique,path):
    if technique == 'greyscale':
        img = cv2.imread(path, 127)
        plt.subplot(1, 2, 1)
        plt.title("Original")
        plt.imshow(img)
        plt.show()
    elif technique == 'brightness':
        image = cv2.imread(path)
        #adjust the brightness and contrast
        brightness = 5
        contrast = 1.5
        img = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)
        plt.title("Brightness & Contrast")
        plt.imshow(img)
        plt.show()
        
    elif technique == 'sharpen':
        image = cv2.imread(path)
        #create a sharpening kernel
        kernel = np.array([[0,-1,0], [-1,5,-1],[0,-1,0]])
        #sharpen the image
        sharpened_image = cv2.filter2D(image, -1, kernel)
        plt.title("Sharpened")
        plt.imshow(sharpened_image)
        plt.show()
    elif technique == 'noise':
        image = cv2.imread(path)
        #filter the image
        noise_image = cv2.medianBlur(image, 15)
        plt.title("Median Blur")
        plt.imshow(noise_image)
        plt.show()
    elif technique == 'scaling':
        image = cv2.imread(path)
        #scaled the image
        scaled_image = cv2.resize(image, None, fx=10, fy=10)
        plt.title("Scaled")
        plt.imshow(scaled_image)
        plt.show()
            
    elif technique == 'inverse':
        image = cv2.imread(path)
        #inverse the image
        inverse_image = 239 - image
        plt.title("Inverse")
        plt.imshow(inverse_image)
        plt.show()
            
    else:
        print("Invalid Input")
```
