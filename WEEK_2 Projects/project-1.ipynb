```python
def get_image(features):
    for name, names in members.items():

        # If we remove 'if key in features', the condition would check ALL keys in names, including "profile"
        # the 'if key in features' check keys that exist in features ("firstname" and "lastname").
        # "profile" is ignored in the condition (so it doesnâ€™t cause failure).
        if all(features.get(key, False) == value   for key, value in names.items() if key in features):
            image_name = names['firstname']
            extensions = ["jpg", "png"]

            for ext in extensions:
                image = cv2.imread(f'img/{image_name}.{ext}')
                if image is not None:  # Check if image is loaded
                    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
                    plt.subplot(1, 2, 1)
                    plt.title("Original")
                    plt.imshow(image)
                    plt.show() #ensures the image is displayed
                    print(names['profile'])
                    return name
    return 'Member does not exist or member does not have an image'
```
