```python
# Function to load and display images
def display_image(path, title):
    image = cv2.imread(path)
    if image is not None:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        plt.imshow(image)
        plt.title(title)
        plt.axis('off')
        plt.show()
        return True
    return False

# edge detection function
def edge_detection_techniques(technique, category, key):
    for ext in extensions:
        path = f'img/{category}/{key}.{ext}'
        image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        if image is None:
            continue
        plt.subplot(1, 2, 1)
        plt.title("Original")
        plt.imshow(image)

        if technique == 'sobel':
            #resize images
            image = cv2.resize(image,(500,400))
            #Apply sobel edge detection
            sobel_x = cv2.Sobel(image, cv2.CV_64F,1,0,ksize=3)
            sobel_y = cv2.Sobel(image, cv2.CV_64F,0,1,ksize=3)
            #Combine the gradient images
            edge_detected_img = cv2.magnitude(sobel_x, sobel_y)


        elif technique == 'prewitt':
            #resize images
            image = cv2.resize(image, (500,400))
            #define prewitt kernels
            kernel_x = np.array([[-1,-1,-1],
                                [0,0,0],
                                [1,1,1]])
            kernel_y = np.array([[-1,0,1],
                                [-1,0,1],
                                [-1,0,1]])
            #Apply prewitt edge detection
            prewitt_x = cv2.filter2D(image, -1, kernel_x)
            prewitt_y = cv2.filter2D(image, -1, kernel_y)
            #compute magnitude of gradients manually
            prewitt_combined = np.sqrt(np.square(prewitt_x) + np.square(prewitt_y))
            edge_detected_img = np.uint8(prewitt_combined)

        elif technique == 'robert':
            #perform gaussian blur
            image = cv2.GaussianBlur(image, (5,5), 0)
            #define robert kernels
            kernel_x = np.array([[1,0],
                                [0,-1]])
            kernel_y = np.array([[0,1],
                                [-1,0]])
            #Apply robert edge detection
            robert_x = cv2.filter2D(image, -1, kernel_x)
            robert_y = cv2.filter2D(image, -1, kernel_y)
            #compute magnitude of gradients manually
            robert_combined = np.sqrt(np.square(robert_x) + np.square(robert_y))
            edge_detected_img = np.uint8(robert_combined)

        elif technique == 'laplacian':
            #resize
            image = cv2.resize(image, (400,400))
            # apply laplacian edge detection
            laplacian = cv2.Laplacian(image, cv2.CV_64F)
            #convert the output to an 8-bit image
            edge_detected_img = cv2.convertScaleAbs(laplacian)

        elif technique == 'canny':
            #resize
            image = cv2.resize(image, (400,400))
            # apply canny edge detection
            edge_detected_img = cv2.Canny(image,100,200)
        
        else:
            print("Invalid edge detection technique.")
            return
        
        plt.subplot(1, 2, 2)
        plt.title("Edge Detected Image")
        plt.imshow(edge_detected_img)
        plt.show()
        return
    print("Image not found.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import re

# Categories dictionary
categories = {
    "Traditional": {"benin": "Benin Head", "gelede": "Gelede Headdress", "nok": "Male NOK Head", "yoruba": "Yoruba Head"},
    "Contemporary": {"home": "The Way Home", "orange": "The Orange Market", "palm": "The Palm Wine Tappers"},
    "Modern": {"dancer": "The Dancer", "iwin": "Iwin", "masks": "Masks"}
}

# Case-insensitive category lookup
type_lookup = {key.lower(): key for key in categories.keys()}

# Supported image extensions
extensions = ["jpg", "png", "jpeg"]

# Main program loop
while True:    
    category = input("Enter your category (Traditional, Contemporary, Modern): ").strip().lower()
    if category not in type_lookup:
        print("Invalid category. Choose from:", list(type_lookup.keys()))
        continue
    category_key = type_lookup[category]
    subcategories = categories[category_key]

    for key, value in subcategories.items():
        for ext in extensions:
            path = f'img/{category_key}/{key}.{ext}'
            if display_image(path, value):
                break
    
    valid_inputs = ['sobel', 'prewitt', 'robert', 'laplacian', 'canny']
    while True:
        technique = input("Enter edge detection technique: ").strip().lower()
        if technique not in valid_inputs:
            print("Invalid technique. Choose from:", valid_inputs)
            continue

        selected_image = input("Enter the name of the selected image: ").strip().lower()
        selected_key = next((k for k, v in subcategories.items() if v.lower() == selected_image), None)
        
        if not selected_key:
            print("Image not found in the selected category.")
            continue
        
        edge_detection_techniques(technique, category_key, selected_key)
        
        if input("Do you want to continue? (Y/N): ").strip().lower() != 'y':
            break
```
