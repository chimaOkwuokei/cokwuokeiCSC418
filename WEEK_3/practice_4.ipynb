```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('img/sst.png',0)

#plot the original image
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img)
                 # tx      ty
cropped_img = img[250:300, 50:100]

#plot the translated image
plt.subplot(1,2,2)
plt.title("Cropped Image")
plt.imshow(cropped_img)

cv.imwrite('img/cropped_out.png', cropped_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
