```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('img/sst.png',0)

#plot the original image
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img)

rows, cols = img.shape
img_rotation = cv.warpAffine(img,cv.getRotationMatrix2D((cols/2,rows/2), 100, 0.6)
                                  , (cols,rows))

#plot the translated image
plt.subplot(1,2,2)
plt.title("Rotated Image")
plt.imshow(img_rotation)

cv.imwrite('img/rotation_out.jpg', img_rotation)
cv.waitKey(0)
cv.destroyAllWindows()
```
