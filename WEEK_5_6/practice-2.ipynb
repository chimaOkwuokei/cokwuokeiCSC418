```python
import cv2
from matplotlib import pyplot as plt

#load the input image
image = cv2.imread('images/friends.png')

#opencv opens image as BRG
#but we want it aas RGB
#convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


#perform face detection
face_data = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

found = face_data.detectMultiScale(gray_image, minSize=(20,20))

amount_found = len(found)

if amount_found != 0:
    #draw the bounding boxes around the detected faces
    for (x, y, w, h) in found:
        cv2.rectangle(rgb_image, (x, y), (x+h, y+w), (0,255,0), 5)
        
#display the output image
plt.subplot(1,1,1)
plt.imshow(rgb_image)
plt.show()
```
